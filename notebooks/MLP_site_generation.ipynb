```python

def get_all_diseases_df(folder):
    sids_vus = set()
    df_total = pd.DataFrame()
    for maladie in ["AD", "ADHD", "BIP", "MCI", "SCHZ", "TBI"]:
        df_raw = pd.read_csv(folder / f"{maladie}_combination_all_metrics_CamCAN.csv.gz")

                # Filtrer les lignes où "aug" est dans le sid
        df_aug = df_raw[df_raw['sid'].astype(str).str.contains('aug')]

        # Afficher les sids uniques correspondants
        sids_aug = df_aug['sid'].unique()
        print(f"Augmented SIDs for {maladie}:")
        print(sids_aug)
        
        # On enlève les SIDs déjà vus
        df_filtré = df_raw[~df_raw["sid"].isin(sids_vus)]
        
        # On ajoute les nouveaux SIDs à notre set
        sids_vus.update(df_filtré["sid"].unique())
        
        # On concatène le DataFrame filtré
        df_total = pd.concat([df_total, df_filtré], ignore_index=True)
    return df_total
```

```python
df = get_all_diseases_df(Path("DONNES_F/COMPILATIONS"))
train_df, test_df = split_train_test(df, test_size=0.5,random_state=42)

os.makedirs("DONNES_MLP", exist_ok=True)
train_df.to_csv("DONNES_MLP/train_data_all.csv", index=False)
test_df.to_csv("DONNES_MLP/evalutation_data_all.csv", index=False)
```
