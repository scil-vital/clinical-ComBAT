# Figure Maker

This notebook builds composite figures from existing distribution analysis images.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_paths = [
    Path('RESULTS/DISTRIBUTION_ANALYSIS/COMPILATION_DISTRIBUTION/TBI/afd/afd_mni_IFOF_L.png'),
    Path('RESULTS/DISTRIBUTION_ANALYSIS/COMPILATION_DISTRIBUTION/AD/fw/fw_mni_AC.png'),
]

fig, axes = plt.subplots(2, 1, figsize=(8, 8))
for ax, img_path in zip(axes, image_paths):
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.axis('off')

fig.suptitle('Pathology-Driven Shifts in Diffusion Metric Distributions', fontsize=16, y=0.98)
plt.tight_layout(pad=0.5, rect=(0, 0, 1, 0.95))
plt.show()
```
